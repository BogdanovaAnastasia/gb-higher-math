## Домашнее задание по теме "Введение в аналитическую геометрию"

**1\.2 Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.**

```python
def vector_sum(vector):
    arg_sum = 0
    for a in vector:
        arg_sum += a**2
    return sqrt(arg_sum)
```

```python
A = array([10, 10, 10])
B = array([0, 0, -10])
```

```python
C = A + B
C
```

```python
vector_sum(A)
```

```python
vector_sum(B)
```

```python
vector_sum(C)
```

## Домашнее задание по теме "Графики на плоскости"

**1\. Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b**

```python
def graph(k, a, b):
    x = linspace(-100, 100, 1000)
    y = k * cos(x - a) + b
    plot(x, y, label=label_construct(k, a, b))
```

```python
def label_construct(k, a, b):
    if a < 0 and b < 0:
        return f'{k}*cos(x + {abs(a)}) - {abs(b)}'
    if a > 0 and b < 0:
        return f'{k}*cos(x - {a}) - {abs(b)}'
    if a < 0 and b > 0:
        return f'{k}*cos(x + {abs(a)}) + {b}'
    if a >= 0 and b >= 0:
        return f'{k}*cos(x - {a}) + {b}'
```

```python
figure(figsize=(15, 5))
graph(1, 0, 0)
graph(5, 2, 1)
graph(2,  -5, -10)
ylim(-15, 15)
grid(True)
legend()
show()
```

**3\.1 Напишите код, который будет переводить полярные координаты в декартовы.**

```python
def polar_to_dekart(rho, pfi):
    return round(rho*cos(pfi), 2), round(rho*sin(pfi), 2)
```

**4\.1 Решите систему уравнений:**  
$\begin{cases}
y = x^{2} - 1\\
exp(x) + x\cdot(1 - y) = 1
\end{cases}$  

```python
def f(p):
    x, y = p
    return (y - x**2 + 1, exp(x) + x - x*y -1)
```

```python
x = linspace(-2, 5, 200)
plot(x, x**2 - 1, label="y = x**2 – 1")
plot(x, ((x + exp(x) - 1)/x), label="exp(x) + x * (1 – y) = 1")
xlabel('x')
ylabel('y')
xlim(-2, 5)
ylim(-1.5, 20)
legend()
grid(True)
show()
```

```python
fsolve(f, (4, 17))
```

```python
fsolve(f, (3, 6))
```

```python
fsolve(f, (-1.5, 2))
```

**4\.2 Решите систему уравнений:**  
$\begin{cases}
y = x^{2} - 1   \\
exp(x) + x\cdot(1 - y) > 1
\end{cases}$  

```python
# exp(x) + x * (1 – y) > 1
# exp(x) + x – x*y > 1
# -x*y > 1 - exp(x) - x
# -y > (1 - exp(x) - x)/x
# y < (x + exp(x) - 1)/x - множество точек под кривой

x = linspace(-10, 10, 200)
x1 = linspace(-1.58, 2.62, 210)
y1 = x1**2 - 1

fill_between(x1, x1**2 - 1, ((x1 + exp(x1) - 1)/x1), color='g', label="exp(x) + x * (1 – y) > 1")
plot(x, x**2 - 1, label="y = x**2 – 1")
plot(x, ((x + exp(x) - 1)/x), label="exp(x) + x * (1 – y) = 1")
plot(x1, y1, 'r', label="y = x**2 – 1, exp(x) + x * (1 – y) > 1")
xlabel('x')
ylabel('y')
xlim(-2, 5)
ylim(-1.5, 20)
legend()
grid(True)
show()
```

```python
fsolve(f, (2.62, 5.85))
```
