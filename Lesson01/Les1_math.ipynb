**4\. Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.**

```python
# Пусть будет k = 1 для одной кривой и k = 5 - для другой
x = np.linspace(-2*np.pi, 2*np.pi, 200)
plt.plot(x, np.cos(x))
plt.plot(x, np.cos(5*x))
plt.legend(['y(x) = cos(x)', 'y(x) = cos(5x)'])
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.6, 1.6) 
plt.grid(True)
plt.show()
```
